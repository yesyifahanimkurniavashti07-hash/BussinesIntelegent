```python
dataset = pd.read_csv('/content/Dataset_Missing_Values.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

1. setelah berhasil menimport data, selanjutnya kita akan mengisikan "Missing Values" =  NAN.
caranya bagaimana?

2. untuk mengisikan missing values kita dapat menggunakan fungction "SIMPLE IMPUTER", untuk memanggil fungction IMPUTER nantinya untuk mengunakan nilai "RATA RATA ATAU MEAN" dari kolomnya menggunakan libary "SKLEARN.IMPUTE" kemudian kita panggil fungctionnya yaitu SIMPLE IMPUTER.

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
```

```python
print (x)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print (x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```
