```python
dataset = pd.read_csv('/content/Dataset_tugas_bi_yesy.csv' , sep=';')
x = dataset.iloc[:, 0:3].values
y = dataset.iloc[:, 3].values
```

- sep=';' berarti pemisah antar kolom adalah titik koma (;), bukan koma (,).

2. x = dataset.iloc[:, 0:3].values
- Kode ini mengambil data kolom ke-0 sampai ke-2 (3 kolom pertama).
- iloc digunakan untuk mengambil data berdasarkan posisi indeks.
- : artinya ambil semua baris
- 0:3 artinya ambil kolom ke-0, 1, dan 2
- .values mengubah data menjadi array NumPy.

3. y = dataset.iloc[:, 3].values
- Kode ini mengambil kolom ke-3 saja (kolom ke-4).
- Digunakan sebagai target atau label yang ingin diprediksi.

Secara singkat, bagian x berisi data input (fitur), yaitu nama, umur, dan penghasilan dari setiap orang. Terlihat bahwa data sudah berbentuk array, dan ada beberapa nilai yang kosong (nan) yang menandakan data tidak tersedia.

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
```

- SimpleImputer dipanggil untuk membuat alat pengisi data kosong dengan strategi rata-rata (mean).
- imputer.fit(x[:, 1:3]) menghitung nilai rata-rata dari kolom ke-1 sampai ke-2 (biasanya umur dan penghasilan).
- imputer.transform(...) mengganti nilai yang kosong (nan) pada kolom tersebut dengan nilai rata-rata yang sudah dihitung.

```python
print (x)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse_output=False), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

Kode ini digunakan untuk mengubah data kategori (teks) menjadi angka agar bisa diproses oleh model machine learning.
- ColumnTransformer dipakai untuk menentukan kolom mana yang akan diubah,
- OneHotEncoder berfungsi mengubah data kategori (misalnya nama pada kolom pertama) menjadi bentuk angka dalam beberapa kolom biner (0 dan 1).
- ('encoder', OneHotEncoder(sparse_output=False), [0]), artinya kolom ke-0 akan di-encode
- remainder='passthrough' berarti kolom lainnya tetap digunakan tanpa perubahan.
- ct.fit_transform(x) menjalankan proses encoding tersebut, lalu hasilnya diubah menjadi array NumPy.

```python
print (x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

Kode tersebut digunakan untuk mengubah data label (kategori) menjadi angka.
- LabelEncoder adalah alat dari sklearn yang berfungsi mengonversi data seperti 'Yes' dan 'No' menjadi bentuk numerik, misalnya:
'No' → 0
'Yes' → 1
- Baris le.fit_transform(y) akan membaca semua nilai pada y, menentukan kategorinya, lalu langsung mengubahnya menjadi angka.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

Kode tersebut digunakan untuk membagi dataset menjadi data latih dan data uji.
- train_test_split berfungsi memisahkan data menjadi dua bagian:
Data training (x_train, y_train) → digunakan untuk melatih model
Data testing (x_test, y_test) → digunakan untuk menguji performa model
- Parameter test_size = 0.3 berarti 30% data digunakan untuk testing, sedangkan 70% sisanya untuk training.
- random_state = 1 digunakan agar hasil pembagian data tetap sama (konsisten) setiap kali kode dijalankan.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:,3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
```

Kode tersebut digunakan untuk menormalkan (menskalakan) data numerik agar memiliki skala yang sama.

- StandardScaler berfungsi mengubah data sehingga memiliki rata-rata 0 dan standar deviasi 1. Ini penting agar model machine learning tidak bias terhadap nilai yang terlalu besar.
- sc.fit_transform(x_train[:, 3:]) → menghitung rata-rata dan standar deviasi dari data training (kolom ke-3 ke atas), lalu langsung mengubah nilainya.
sc.transform(x_test[:, 3:]) → menggunakan perhitungan dari data training tadi untuk mengubah data testing (tanpa menghitung ulang).
